```python
import json

from pylab import rcParams
rcParams["figure.figsize"] = 15,7

import numpy as np

import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

import stumpy

import pandas as pd

from matplotlib.pyplot import text
```

```python
with open('/home/juanjops/Projects/time_series/heart_beat/files/HR2019-04-10.json', 'r') as f:
    array = json.load(f)

array["activities-heart-intraday"]["dataset"]
```

```python
dataset = pd.DataFrame(array["activities-heart-intraday"]["dataset"])
```

```python
dataset.head(10)
```

```python
dataset.shape
```

```python
dataset = dataset.set_index(pd.to_datetime(dataset["time"]))
dataset = dataset.drop(columns=["time"])
dataset = dataset.resample("1min").mean()
dataset = dataset.asfreq("1min")
dataset = dataset.interpolate()
```

```python
dataset["indice"] = range(0, dataset.shape[0])
```

```python
dataset.value.plot()
```

```python
dataset.shape
```

# Matrix Profile

```python
m = 60
mp = stumpy.stump(dataset['value'], m)
```

```python
mp
```

```python
dataset_filter = dataset.head(len(mp[:,0]))
dataset_filter["matrix_profile"] = mp[:,0]
```

```python
level_pair_importance = 0
point_1, point_2 = np.unique(sorted(mp, key=lambda a_entry: a_entry[0])[level_pair_importance][1:4])
```

```python
point_1, point_2
```

```python
fig, axs = plt.subplots(2, sharex=False, gridspec_kw={'hspace': 0})
plt.suptitle('Motif (Pattern) Discovery', fontsize='30')

axs[0].plot(dataset_filter['value'].values)
axs[0].set_ylabel('Values', fontsize='20')
rect = Rectangle((point_1, 0), m, 160, facecolor='lightgrey')
axs[0].add_patch(rect)
rect = Rectangle((point_2, 0), m, 160, facecolor='lightgrey')
axs[0].add_patch(rect)
axs[1].set_xlabel('Time', fontsize ='20',)
axs[1].set_ylabel('Matrix Profile', fontsize='20')
axs[0].add_patch(rect)
axs[1].set_xlabel('Time', fontsize ='20',)
axs[1].set_ylabel('Matrix Profile', fontsize='20')
axs[1].plot(dataset_filter['matrix_profile'])
axs[1].axvline(dataset.index[point_1], linestyle="dashed")
axs[1].axvline(dataset.index[point_2], linestyle="dashed")
plt.show()
```

# Pattern Search

```python
pattern = dataset['value'].loc["2021-01-27 08:31:00":"2021-01-27 09:30:00"]
```

```python
plt.plot(pattern)
```

```python
distance_profile = stumpy.core.mass(pattern.values, dataset["value"])
```

```python
pattern_point, similar_point = dataset.loc["2021-01-27 08:31:00"].indice, np.argpartition(distance_profile, 1)[1]
```

```python
fig, axs = plt.subplots(1, sharex=False, gridspec_kw={'hspace': 0})
plt.suptitle('Search Pattern', fontsize='30')

axs.plot(dataset['value'].values)
axs.set_ylabel('Values', fontsize='20')
rect = Rectangle((pattern_point, 0), m, 160, facecolor='lightgrey')
axs.add_patch(rect)
rect = Rectangle((similar_point, 0), m, 160, facecolor='lightgrey')
axs.add_patch(rect)
axs.add_patch(rect)

axs.text(pattern_point, 20, "pattern")
axs.text(similar_point, 20, "best_similar_pattern")

plt.show()
```
